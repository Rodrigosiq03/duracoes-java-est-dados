```python
import matplotlib.pyplot as plt
import pandas as pd

mediasbubble = [2136.42, 8457.52, 34390.82, 137359.82, 554080.55, 2227754.4, 8637437.8]
labels = ['40000', '80000', '160000', '320000', '640000', '1280000', '2560000']
df = pd.DataFrame({'medias': mediasbubble}, index=labels)
ax = df.plot.bar(rot=0, color='blue', legend=False, figsize=(10, 5))
ticks = [0, 2000000, 4000000, 6000000, 8000000]
labels = ['0', '2M', '4M', '6M', '8M']
ax.set_yticks(ticks)
ax.set_yticklabels(labels)
for i, v in enumerate(mediasbubble):
    ax.annotate(str(v), xy=(i, v), ha='center', va='bottom')
    
plt.title('Tempo médio de execução para cada tamanho de vetor em BubbleSort')
plt.xlabel('Tamanho do vetor')
plt.ylabel('Tempo médio de execução (ms)')

plt.savefig('bubble.png', dpi=300, bbox_inches='tight')

```
